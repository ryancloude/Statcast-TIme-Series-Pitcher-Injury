```python
import pandas as pd
import re
#import unidecode
import numpy as np
import matplotlib.pyplot as plt
import datetime
from google.colab import drive
from tqdm.notebook import tqdm
import unidecode
tqdm.pandas()
drive.mount('/content/drive')
```

```python
bio = pd.read_csv('/content/drive/MyDrive/data_490/People.csv')
```

```python
bio = bio[['birthYear', 'birthMonth', 'birthDay','nameFirst', 'nameLast', 'nameGiven','weight', 'height']]
bio
```

```python
bio['player_name'] = bio.apply(lambda row: f'{row.nameFirst} {row.nameLast}', axis=1)
bio 
```

```python
#Combines columns birthYear, BirthMonth, and BirthDay to create birth_date
birth_date = bio[['birthYear','birthMonth','birthDay']].copy()
birth_date.columns = ['year', 'month', 'day']
birth_date = pd.to_datetime(birth_date)
bio['birth_date'] = birth_date
```

```python
bio = bio[['birth_date', 'height', 'weight', 'player_name']]
bio = bio.drop_duplicates(subset='player_name')
bio
```

```python
df = pd.read_csv('/content/drive/MyDrive/data_490/processed_data/pandas_30_day')
```

```python
df.player_name = df.player_name.astype(str)
```

```python
df.player_name = df.player_name.apply(unidecode.unidecode)
```

```python
df = df.merge(bio, on=['player_name'], how='left')
```

```python
df['player_age'] = pd.to_datetime(df.game_date) - df.birth_date
```

```python
df = df[['player_name', 'game_date','weight','height','player_age']]
```

```python
df.to_csv('/content/drive/MyDrive/data_490/processed_data/bio_data.csv', index=False)
```
